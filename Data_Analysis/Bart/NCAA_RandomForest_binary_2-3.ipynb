```python
import pandas as pd
# import matplotlib as plt
# from sklearn.datasets import make_blobs
import sklearn as skl
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# import tensorflow as tf
from pathlib import Path  
from sklearn.ensemble import RandomForestClassifier
```

```python
mm_teams_merged_df = pd.read_csv("../../../Module 20 NCAA stats/mm_teams_merged_2011-21.csv")
mm_teams_merged_df.head()
```

### convert tournament wins (0-6) into binary columns (0 or 1);

```python
def First_Wins(row):
    if row > 0:
        result = '1'
        
    else:
        result = '0'

    return result

mm_teams_merged_df['First_Round_Wins'] = mm_teams_merged_df['Tournament Wins'].apply(First_Wins)
```

```python
def Second_Wins(row):
    if row > 1:
        result = '1'
    else:
        result = '0'
    return result

mm_teams_merged_df['Second_Round_Wins'] = mm_teams_merged_df['Tournament Wins'].apply(Second_Wins)
```

```python
def Third_Wins(row):
    if row > 2:
        result = '1'
    else:
        result = '0'
    return result

mm_teams_merged_df['Sweet_Sixteen_Wins'] = mm_teams_merged_df['Tournament Wins'].apply(Third_Wins)
```

```python
def Fourth_Wins(row):
    if row > 3:
        result = '1'
    else:
        result = '0'
    return result

mm_teams_merged_df['Elite_Eight_Wins'] = mm_teams_merged_df['Tournament Wins'].apply(Fourth_Wins)
```

```python
def Fifth_Wins(row):
    if row > 4:
        result = '1'
    else:
        result = '0'
    return result

mm_teams_merged_df['Final_Four_Wins'] = mm_teams_merged_df['Tournament Wins'].apply(Fifth_Wins)
```

```python
def Sixth_Wins(row):
    if row > 5:
        result = '1'
    else:
        result = '0'
    return result

mm_teams_merged_df['Championship_Wins'] = mm_teams_merged_df['Tournament Wins'].apply(Sixth_Wins)
mm_teams_merged_df.head(10)
```

```python
# Export dataframe
# all_rnds_df_path = Path("../../../Module 20 NCAA stats/mm_teams_2011-21_rnds.csv")
# mm_teams_merged_df.to_csv(all_rnds_df_path, index = False)  
```

### Define features and target for round_1 model training

```python
# Define the features set.
X_rnd1 = mm_teams_merged_df.drop(columns= ["Team", "Tournament Wins", 'First_Round_Wins', 'Second_Round_Wins', 'Sweet_Sixteen_Wins', 'Elite_Eight_Wins', 'Final_Four_Wins', 'Championship_Wins'])
X_rnd1.head()
```

```python
# Define the target set.
y_rnd1 = mm_teams_merged_df['First_Round_Wins']
y_rnd1.head()
```

### Split data

```python
# Splitting into Train and Test sets.
X_train_rnd1, X_test_rnd1, y_train_rnd1, y_test_rnd1 = train_test_split(X_rnd1, y_rnd1, random_state=78)
X_test_rnd1.head(10)

# stratify=y?
```

```python
X_test_rnd1.info()
# X_test_rnd1.shape()
```

### scale data

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler_rnd1 = scaler.fit(X_train_rnd1)

# Scaling the data.
X_train_scaled_rnd1 = X_scaler_rnd1.transform(X_train_rnd1)
X_test_scaled_rnd1 = X_scaler_rnd1.transform(X_test_rnd1)
```

```python
# X_train_scaled_rnd1
```

### initiate and fit RandomForestClassifier model

```python
# Create a random forest classifier. keep n_estimators between 64-128; 
# changing to 500 changes results very little
rf_model_rnd1 = RandomForestClassifier(n_estimators=128, random_state=78) 
```

```python
# Fitting the model
rf_model_rnd1 = rf_model_rnd1.fit(X_train_scaled_rnd1, y_train_rnd1)
```

### make predictions

```python
# Making predictions using the testing data.
predictions_rnd1 = rf_model_rnd1.predict(X_test_scaled_rnd1)
```

```python
predictions_rnd1
```

### calculate model metrics

```python
# Calculating the confusion matrix.
cm_rnd1 = confusion_matrix(y_test_rnd1, predictions_rnd1)
# confusion_matrix(y_test, predictions).ravel()
# Create a DataFrame from the confusion matrix.
# cm_df = pd.DataFrame(
#     cm, index=["Actual 0", "Actual 1", "Actual 2", "Actual 3", "Actual 4", "Actual 5", "Actual 6"], columns=["Predicted 0", "Predicted 1", "Predicted 2","Predicted 3", "Predicted 4","Predicted 5", "Predicted 6"])
cm__rnd1_df = pd.DataFrame(
    cm_rnd1, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])
cm__rnd1_df
```

```python
# Calculating the accuracy score.
acc_score_rnd1 = accuracy_score(y_test_rnd1, predictions_rnd1)
acc_score_rnd1
```

```python
# Displaying results
print("Confusion Matrix Round 1")
# display(cm_df)
display(cm_rnd1)
print(f"Accuracy Score Round 1 : {acc_score_rnd1}\n")
print("Classification Report Round 1")
print(classification_report(y_test_rnd1, predictions_rnd1))
```
