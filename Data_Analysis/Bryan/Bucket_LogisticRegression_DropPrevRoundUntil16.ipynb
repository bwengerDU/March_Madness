```python
import pandas as pd
```

```python
data_df = pd.read_csv('/Users/bryanwenger/Desktop/Final_Project/March_Madness/Data_Creation/Bryan/mm_teams.csv')
data_df.head(10)
```

```python
data_df.drop(columns=['Team', 'Conference'], inplace=True)
data_df.head()
```

```python
data_df.count()
```

```python
y = data_df["Tournament_Wins"]
X = data_df.drop(columns="Tournament_Wins")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=2000,
                                random_state=7)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
# Convert data frame in which tournament wins are binary 0=0 1-6=1
data_df.Tournament_Wins[data_df.Tournament_Wins == '1'] = 1
data_df.Tournament_Wins[data_df.Tournament_Wins == '2'] = 1
data_df.Tournament_Wins[data_df.Tournament_Wins == '3'] = 1
data_df.Tournament_Wins[data_df.Tournament_Wins == '4'] = 1
data_df.Tournament_Wins[data_df.Tournament_Wins == '5'] = 1
data_df.Tournament_Wins[data_df.Tournament_Wins == '6'] = 1
data_df.Tournament_Wins[data_df.Tournament_Wins == '0'] = 0
data_df.head(20)
```

```python
def First_Wins(row):


    if row > 0:


        result = '1'
        

    else:


        result = '0'


    return result


 


data_df['First_Round_Wins'] = data_df['Tournament_Wins'].apply(First_Wins)
```

```python
data_df.head(10)
```

```python
def Second_Wins(row):
    if row > 1:
        result = '1'
    else:
        result = '0'
    return result

data_df['Second_Round_Wins'] = data_df['Tournament_Wins'].apply(Second_Wins)
data_df.head(3)
```

```python
def Third_Wins(row):
    if row > 2:
        result = '1'
    else:
        result = '0'
    return result

data_df['Sweet_Sixteen_Wins'] = data_df['Tournament_Wins'].apply(Third_Wins)
data_df.head(3)
```

```python
def Fourth_Wins(row):
    if row > 3:
        result = '1'
    else:
        result = '0'
    return result

data_df['Elite_Eight_Wins'] = data_df['Tournament_Wins'].apply(Fourth_Wins)
data_df.head(10)
```

```python
def Fifth_Wins(row):
    if row > 4:
        result = '1'
    else:
        result = '0'
    return result

data_df['Final_Four_Wins'] = data_df['Tournament_Wins'].apply(Fifth_Wins)
data_df.head(3)
```

```python
def Sixth_Wins(row):
    if row > 5:
        result = '1'
    else:
        result = '0'
    return result

data_df['Championship_Wins'] = data_df['Tournament_Wins'].apply(Sixth_Wins)
data_df.head(30)
```

```python
data_df.to_csv(r'/Users/bryanwenger/Desktop/Final_Project/March_Madness/Data_Creation/Bryan/wins_by_round.csv')
```
